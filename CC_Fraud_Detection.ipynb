```python
#import de bibliotecas e ferramentas/scripts de machine learning

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
```

## Iremos extrair o arquivo csv para nossas análises e treinamento do script diretamente do repositório em meu GitHub. Mais informações sobre o data set estarão disponíveis no arquivo readme.md

```python
df = pd.read_csv('https://raw.githubusercontent.com/viaNNation/CC_Fraud_Detection/main/creditcard3.csv')
df
```

```python
#visualização das informaçoes basicas do df
df.info()
```

## Nenhuma linha do DF possui valores nulos, e temos variáveis float e int.



```python
#outra forma de conferir

df.isnull().sum()
```

```python
#verificando se temos linhas duplicadas no data frame, e já retirando as mesmas logo em seguida

df.duplicated().sum()
```

```python
#como há linhas duplicadas, iremos retirá-las para melhor precisão dos scripts e modelos aplicados

df = df.drop_duplicates()
```

```python
df['Class'] = df['Class'].astype(int)
df
```

```python
df.describe()
```

## Agora temos um data frame trabalhado sem nenhuma linha duplicada, sem valores nulos, e com a variável class transformada em inteiros. Vamos avaliar a disparidade entre os dados de class 0 e 1


```python
#Verificando quantas linhas do DF possuem o valor de class = 0, indicando que não houve fraude

df_nf = df.Amount[df.Class == 0]
df_nf.describe()
```

## 7999 registros do DF não foram identificados como tentativas de fraude

```python
#Verificando quantas linhas do DF possuem o valor de class = 1, indicando que houve fraude

df_f = df.Amount[df.Class == 1]
df_f.describe()
```

## 473 registros do DF foram identificados como tentativas de fraude
